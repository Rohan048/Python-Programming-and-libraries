```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
xpts = np.array([0,10])
ypts = np.array([0,125])
plt.plot(xpts,ypts)
plt.show()
```

```python
# plot a line with DataFrame (using import pandas)------------------------
Data=({
    "Name":['Rohan','Rahul','rishu','sonu','amol'],
    "Age":[20,19,21,18,24],
    "Weight":[55,50,65,60,75]
})
df = pd.DataFrame(Data)
plt.plot(df["Name"], df["Weight"])
plt.show()
```

```python
# MATPLOTLIB - ADD GRID LINES-----------------------------(grid())
Data=({
     "Name":['Rohan','Rahul','rishu','sonu','amol'],
     "Age":[20,19,21,18,24],
     "Weight":[55,50,65,60,75]
 })
df=pd.DataFrame(Data)
plt.plot(df["Age"], df["Weight"])
plt.grid()
plt.show()
```

```python
# ADD LABELS TO A PLOT-------------
# xlabel() for x-coordinates and ylabel() for y-coordinates
Data=({
      "Name":['Rohan','Rahul','rishu','sonu','amol'],
      "Age":[20,19,21,18,24],
      "Weight":[55,50,65,60,75]
})
df=pd.DataFrame(Data)
plt.plot(df["Age"], df["Weight"])
plt.xlabel("User Age(year)")
plt.ylabel("User Weight(kg)")
plt.show()
```

```python
# PLOT TITLES AND POSITION THEM---------------------
Data=({
      "Name":['Rohan','Rahul','rishu','sonu','amol'],
      "Age":[20,19,21,18,24],
      "Weight":[55,50,65,60,75]
})
df=pd.DataFrame(Data)
#Plot a line graph
plt.plot(df["Age"], df["Weight"])
# lables
plt.xlabel("User Age(year)")
plt.ylabel("User Weight(kg)")
# set the titles
plt.title("Age depends on Weight",loc='right')
#Display the figure
plt.show()
```

```python

# ADD A MATPLOTLIB LENGEND IN GRAPH---------------
a = np.arange(5)
b = [2, 3, 4, 5, 6]
c = [5, 6, 7, 8, 9]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a, b, 'k--', label='Frequency')  #k-- = marker set
ax.plot(a, c, 'k:', label='Periods')

l = ax.legend(loc='lower right')
l.get_frame().set_facecolor('red')

plt.title("Frequency of a single")

plt.show()
```

```python
# Change the font size of the legends-----------------
a = np.arange(5)
b = [2, 3, 4, 5, 6]
c = [5, 6, 7, 8, 9]
fig = plt.figure()
ax = plt.subplot()
ax.plot(a, b, 'k--', label='Frequency')  #k-- = marker set
ax.plot(a, c, 'k:', label='Periods')
l=ax.legend(loc='lower right', fontsize='xx-large')
l.get_frame().set_facecolor('red')
plt.title("Frequency of a single")
plt.show()
```

```python
# PLOT A BAR GRAPH IN MATPLOTLIB--------------

student = np.array(["Rohan", "Rahul", "Karan", "Ramu", "Raju"])
marks = np.array([64, 54, 56, 77, 65])
plt.bar(student,marks)
plt.xlabel("student")
plt.ylabel("marks")
plt.title("students marks in Mathematics")
plt.show()
```

```python
# PLOT A PIE CHART IN MATPLOTLIB-----------------

cricketer = np.array(["Sky", "Rohit", "Sachin", "Virat", "Rana"])
runs = np.array([30, 45, 67, 43, 56])
plt.pie(runs, labels=cricketer, autopct="%1.2f")
plt.title("Runs scored in match")
plt.show()
```

```python
# LINE GRAPH IN MATPLOTLIB-------------

month = np.array(["jan", "feb", "march", "april", "may"])
rain = np.array([50, 49, 59, 39, 48])
plt.plot(month, rain)
plt.xlabel("months")
plt.ylabel("Rainfall(mm) ")
plt.show()
```

```python
# HISTOGRAM IN MATPLOTLIB-------------

arr = np.array([23, 45, 46, 98, 56, 75, 82, 18,49,20])
plt.hist(arr, bins=[0, 20, 40, 60, 80, 100],color='g')
plt.xlabel("marks")
plt.ylabel("students")
plt.title("Students marks ")
plt.show()
```

```python
# SCATTER PLOT IN MATPLOTLIB----------------------

team1_Score = [24, 34, 56, 43, 22, 76, 42]
team2_Score = [94, 44, 43, 32, 65, 41, 10]

scoreRange = [5, 10, 15, 20, 25, 30, 35]

plt.scatter(team1_Score,scoreRange, color='r')
plt.scatter(team2_Score,scoreRange, color='g')

plt.xlabel("Team Score")
plt.ylabel("Score Range")

plt.title("Teams Scored runs in match")
plt.show()
```
